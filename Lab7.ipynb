## Реализация перцептрона с алгоритмом обратного распространения ошибки ручками

```python
class MLP():
    def __init__(self, count_neurons_in_layer, activation_function_in_layer, model="clf"):
        self.neural_network = []
        self.count_neurons_in_layer = count_neurons_in_layer
        self.activation_function_in_layer = activation_function_in_layer
        self.model = model
        self.build_network()
    
    def build_network(self):
        for i in range(len(self.count_neurons_in_layer) - 1):
            self.perceptron(self.count_neurons_in_layer[i], self.count_neurons_in_layer[i + 1])
        
    def perceptron(self, n_inputs, n_outputs):
        weights = self.scale_initiation(n_inputs, n_outputs)
        bias = np.random.normal(size=(1, n_outputs))
        self.neural_network.append({"weights" : weights, "bias" : bias})
        
    def scale_initiation(self, height, width):
        return np.random.normal(size=(height, width))

    def neuron_output(self, weights, inputs, activation):
        return activation(np.dot(inputs, weights))


    def output_values(self, input_values, layer, activation):    
        self.neural_network[layer]["output"] = activation(
            np.dot(input_values, self.neural_network[layer]["weights"]) + self.neural_network[layer]["bias"])

    def error_calculation(self, x, y = []):
        if self.model == "clf":
            return x * (1.0 - x)
        elif self.model == "reg":
            return self.mse(x, y)

    def entropy(self, y_pred, y):
        return -np.sum(np.log(y_pred) * y)
    
    def mse(self, y_pred, y):
        return np.sum(np.power(y_pred - y, 2))

    def delta_calculation(self, y):
        for i in reversed(range(len(self.neural_network))):
            layer = self.neural_network[i]
            if i == len(self.neural_network) - 1:
                errors = np.array(y).reshape((1, self.unique_value)) - layer["output"]
                layer["delta"] = np.multiply(errors, self.error_calculation(layer["output"]))
            else:
                errors = np.dot(self.neural_network[i + 1]["delta"], self.neural_network[i + 1]["weights"].T)
                layer["delta"] = errors * self.error_calculation(layer["output"], y)
                layer["delta_bias"] = np.sum(layer["delta"])

    def weight_update(self, learning_rate=0.1):
        for i in range(len(self.neural_network)):
            layer = self.neural_network[i]
            if i > 0:
                layer["weights"] += learning_rate * np.dot(self.neural_network[i - 1]["output"].T, layer["delta"])
                if i != len(self.neural_network) - 1:
                    layer["bias"] += learning_rate * layer["delta_bias"]
                    
    def onehot(self, y, outputs):
        y_onehot = []
        for i in y:
            one_hot = np.zeros((1, outputs))
            one_hot[0, i] = 1
            y_onehot.append(one_hot)
        return y_onehot

    def predict(self, X_test):
        y_pred = []
        for i in X_test:
            for k in range(len(self.count_neurons_in_layer) - 1):
                if k == 0:
                        self.output_values(i, k, self.activation_function_in_layer[k])
                else:
                        self.output_values(self.neural_network[k - 1]["output"], k, self.activation_function_in_layer[k])
            y_pred.append(np.argmax(self.neural_network[len(self.neural_network) - 1]["output"]))
        return y_pred
    
    def fit(self, X_train, y_train, epochs=1): # тренировка модели
        loss = 10**10
        neural_network_best_param = 0
        if self.model == "clf":
            self.unique_value = len(np.unique(y_train))
        elif self.model == "reg":
            self.unique_value = 1

        for j in range(epochs):
            for i in range(len(X_train)):
                for k in range(len(self.count_neurons_in_layer) - 1):
                    if k == 0:
                        self.output_values(X_train[i, :], k, self.activation_function_in_layer[k])
                    else:
                        self.output_values(self.neural_network[k - 1]["output"], k, self.activation_function_in_layer[k])
                
                y_pred = self.neural_network[len(self.neural_network) - 1]["output"]
                if self.model == "clf":
                    y = self.onehot(y_train[i:i+1], self.unique_value)
                    ent = self.entropy(y_pred, y)
                elif self.model == "reg":
                    y = y_train[i:i+1]
                    ent = self.mse(y_pred, y)
                if loss > ent:
                    loss = ent 
                    neural_network_best_param = self.neural_network.copy() # сохранение лучших весов сети
                #print() 
                self.delta_calculation(y) # вычисление ошибки для каждого веса
                self.weight_update() # обновление веса
        return neural_network_best_param
```

```python
# сигмоида
def sigma(x): # функция активации
    return 1 / (1 + np.exp(-x))

# определение выроятность принадлежность к каждому классу
def softmax(output):
    return np.exp(output) / np.sum(np.exp(output))
```

```python
mlp = MLP(count_neurons_in_layer=[25, 64, 128, 64, 2], activation_function_in_layer=[sigma, sigma, sigma, sigma, softmax])
```

```python
mlp.fit(np.array(X_train_clf), np.array(y_train_clf))
```

```python
y_pred = mlp.predict(np.array(X_test_clf))
```

```python
def accuracy(y_test, y_pred):
    matrix = y_test == y_pred
    return np.float64(np.sum(matrix) / len(matrix))
```

```python
accuracy(y_test_clf, y_pred)
```

```python
def f(x):
    return x

def elu(x):
    for i in range(x.shape[0]):
        for j in range(x.shape[1]):
            if x[i][j] <= 0:
                x[i][j] = np.exp(x[i][j]) - 1
    return x
```

```python
mlp = MLP(count_neurons_in_layer=[25, 64, 128, 64, 1], activation_function_in_layer=[elu, elu, elu, elu, f], model="reg")
```

```python
mlp.fit(np.array(X_train_reg), np.array(y_train_reg), epochs=2)
```

```python
y_pred = mlp.predict(np.array(X_test_reg))
```

```python
def mse(y_pred, y):
    return round(np.mean(np.square(y_pred - y)), 2)
```
